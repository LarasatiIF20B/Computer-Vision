### Case Study Description
As of July 2023, both Apple and Samsung lead the smartphone industry worldwide, with a combined 52.61% of the total market share [(ref1)](https://www.oberlo.com/statistics/smartphone-market-share). In this case, you are challenged to replicate the concept (brighten dark-toned photos), and then compare the result with another approach, i.e., **Contrast Limited Adaptive Histogram Equation (CLAHE)**.

### Import libraries

```python
import numpy as np, cv2
from skimage.io import imsave
from skimage.measure import block_reduce
from torch.nn import MaxPool2d
from torch import from_numpy
from matplotlib import pyplot as plt
from matplotlib.image import imread
```

### User-defined Functions (UDFs)

```python
def plot(
    imgs : list,
    titles : list = ["ORIGINAL", "EDITED"],
    figsize : tuple = (8, 5),
    cmaps : list = [None, None],
    *args, **kwargs
) -> None:

    if len(imgs) > 3:
        raise ValueError("Max. total image = 3.")

    if len(imgs) != len(titles):
        raise ValueError("'imgs' should be in the same length with 'titles'.")

    my_figure = plt.figure(figsize=figsize)
    # NOTE: figsize=(width/horizontally, height/vertically)

    fig_ids = [121, 122] if len(imgs) == 2 else [131, 132, 133]

    for idx, img, title, c in zip(fig_ids, imgs, titles, cmaps):
        m = my_figure.add_subplot(idx)
        plt.imshow(img, cmap=c)
        plt.axis("off")
        plt.title(title)

    plt.tight_layout()
```

```python
def apply_clahe_rgb(img, clipLimit=2, tileGridSize=(10, 10)):
    # LAB color model where L = Lightness, a = green-red, b = blue-yellow
    light, a, b = cv2.split(cv2.cvtColor(img, cv2.COLOR_BGR2LAB)) # convert to LAB color model
    light_clahe = cv2.createCLAHE(clipLimit=clipLimit, tileGridSize=tileGridSize).apply(light) # apply CLAHE for lightness (L) component only
    return cv2.cvtColor(cv2.merge((light_clahe, a, b)), cv2.COLOR_LAB2BGR) # return with reconvert to BGR color model
```

### Warm Up: Image processing (exploration using OpenCV)

* Color image -> Grayscale image

```python
# original image
img_lena = cv2.cvtColor(lena, cv2.COLOR_BGR2RGB) # TODO: Change the color channel from BGR to RGB using cv2.COLOR_BGR2RGB

# color image to grayscaled image
img_grayscaled = cv2.cvtColor(img_lena, cv2.COLOR_RGB2GRAY) # TODO: Change the color channel from BGR to Grayscale using cv2.COLOR_BGR2GRAY
img_grayscaled_rgb = cv2.cvtColor(img_grayscaled, cv2.COLOR_GRAY2RGB)

# grayscaled image to binary image
_, img_binary = cv2.threshold(img_grayscaled, 128, 255, cv2.THRESH_BINARY) # TODO: Change the color channel from Grayscale to Binary using cv2.THRESH_BINARY

# plot both
plot(imgs=[img_lena, img_grayscaled_rgb, img_binary], titles=["ORIGINAL", "GRAYSCALE", "BINARY"], cmaps=[None, None, "gray"])
plt.show()
```

* Check their histograms

```python
# TODO: Plot the histogram from the original image
# Function to plot the histogram
def plot_histogram(image, title):
    # Calculate histogram
    hist, bins = np.histogram(image.ravel(), 256, [0, 256])
    # Plot histogram
    plt.plot(hist, color='black')
    plt.xlim([0, 256])
    plt.title(title)
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.grid(True)

# Plot the histogram for the original image
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.imshow(img_lena)
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plot_histogram(cv2.cvtColor(img_lena, cv2.COLOR_RGB2GRAY), 'Histogram - Original Image')

# Display the plots
plt.tight_layout()
plt.show()
```

### Max Pooling (using Scikit-image)

```python
BLOCK_SIZE = (4, 4, 1)
```

```python
img_edited = block_reduce(img, block_size=BLOCK_SIZE, func=np.max)
plot(imgs=[img, img_edited])
```

### Max Pooling (using PyTorch)

```python
# convert numpy-based image to torch.Tensor type
img_torch = from_numpy(img).permute(2, 0, 1).float()

# prepare Max Pooling operation from PyTorch
max_pooling = MaxPool2d(kernel_size=9, stride=2)

# perform Max Pooling operation
img_torch_edited = max_pooling(img_torch).permute(1, 2, 0).int().numpy()
plot(imgs=[img, img_torch_edited])
```

### Min Pooling and Average Pooling

```python
img_edited_min = block_reduce(lena, block_size=BLOCK_SIZE, func=np.min) # TODO: Recall the block_reduce(), but with func=np.min
plot(imgs=[lena, img_edited_min])
plt.show()
```

```python
img_edited_avg = block_reduce(lena, (2, 2, 1), func=np.mean).astype(int) # TODO: Recall the block_reduce().astype(int), but with func=np.mean
plot(imgs=[lena, img_edited_avg], titles=["ORIGINAL", "EDITED (avg)"])
```

```python
# QUESTIONS
# TODO: What is the difference between Min Pooling and Average Pooling operations?
# If you cannot see the difference, try with another image or change the BLOCK_SIZE.
```

[ ANSWER HERE ]

- Min Pooling mengambil nilai terkecil, sedangkan Average Pooling mengambil rata-rata nilai. Contoh dalam penggunaannya Jika kita ingin mendeteksi tepi yang tajam dalam gambar, Min Pooling mungkin lebih sesuai.
- Min Pooling cenderung mempertahankan fitur yang lebih tajam atau intensitas yang lebih rendah dalam citra, sementara Average Pooling cenderung menghasilkan citra yang lebih halus. contoh dalam penggunaannya Jika kita ingin meratakan gambar atau mendapatkan gambar dengan tingkat keseragaman yang lebih tinggi, Average Pooling dapat lebih sesuai.
- Dalam gambar lena tersebut yang lebih unggul menggunakan average karena menghasilkan gambar yang lebih jelas dan lebih terang dibandingkan dengan Min Pooling. Ketika menggunakan Min Pooling gambar yang dihasilkan lebih tajam namun sedikit pecah dan tidak jelas (blur) dan juga warna lebih tajam karena gelap.

### CLAHE
Now imagine you have a better approach to compete Apple's Photonic Engine and Samsung's Adaptive Tetra-squared Pixel technologies using **Contrast Limited Adaptive Histogram Equalization (CLAHE)**. How's the result?

```python
img_clahe = apply_clahe_rgb(img) # TODO: Call the function apply_clahe_rgb() and pass the variable image of "photo1.jpeg"
plot(imgs=[img, img_edited, img_clahe], titles=["ORIGINAL", "MAX-POOLING", "CLAHE"], figsize=(12, 5), cmaps=[None]*3)
plt.show()
```

```python
# QUESTIONS
# TODO: What are the advantages of using CLAHE (instead of Max Pooling) to brighten the dark-toned image?
```

[ ANSWER HERE ]

  Penggunaan CLAHE (Contrast Limited Adaptive Histogram Equalization) memiliki keunggulan dibandingkan dengan Max Pooling dalam mencerahkan gambar yang memiliki nada gelap:

  - CLAHE adalah metode yang adaptif, yang berarti ia dapat secara otomatis menyesuaikan tingkat penerangan berdasarkan kondisi lokal pada citra. Ini memungkinkannya untuk lebih baik menangani variasi tingkat kecerahan pada berbagai bagian citra.
  - CLAHE cenderung mempertahankan rincian pada citra, bahkan ketika mencerahkan gambar. Hal ini penting terutama jika ada rincian penting yang perlu dipertahankan meskipun dalam kondisi pencahayaan yang minim.
  - CLAHE membatasi kontras citra, sehingga hasilnya tetap optimal tanpa mengorbankan kontras yang dapat penting dalam interpretasi gambar.
  - CLAHE dirancang untuk mengatasi tantangan citra dengan pencahayaan yang rendah atau tidak merata, yang sering terjadi pada gambar-gambar dengan kondisi pencahayaan yang suboptimal.
  - CLAHE memberikan hasil visual yang lebih baik dalam mencerahkan gambar dibandingkan dengan metode sederhana seperti Max Pooling, terutama ketika citra memiliki variasi tingkat kecerahan yang signifikan.
